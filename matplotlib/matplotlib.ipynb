## Matplotlib example workflow

```python
# 0- import matloptlib

# 1- prepare data
x= [1,2,3,4]
y=[11,22,33,44]

# 2- Setup plot
fig, ax = plt.subplots(figsize = (10, 10)) #width, height

# 3- Plot Data
ax.plot(x, y)

# 4- Customize plot
ax.set(title= "Simple plot",
      xlabel= "x-axis",
      ylabel= "y-axis")

# 5- Save and show(whole figure)
fig.savefig("./simple-plot.png")
```

```python
import numpy as np
```

```python
#create some data
x= np.linspace(0,10,100)
x
```

```python
#plot the data & create a line plot(default)
fig, ax = plt.subplots()
ax.plot(x, x**2)
ax.set(title ="Line Plot",
      xlabel = "x",
      ylabel= "x^2")
```

```python
#Scatter Plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
ax.set(title= "Scatter Plot",
      xlabel = "x",
      ylabel = "e^x")
```

```python
#aother scatter plot
fig,ax = plt.subplots()
ax.scatter(x, np.sin(x))
ax.set(title= "Another Scatter Plot",
      xlabel= "x",
      ylabel = "sin(x)")
```

```python
# Make  a plot from a dictionary
nut_butter_prices = {
    "Almond Butter": 10,
    "Peanut Butter": 8,
    "Cashew  Butter": 12
}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Bar Graph",
      xlabel= "Nut Butters",
      ylabel= "Prices");
```

```python
#bar horizontal
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title="Horizontal Bar Graph",
      xlabel= "Prices",
      ylabel= "Nut Butters");
```

```python
#Histograms
#create data for histograms
x= np.random.randn(1000) #gives 1000 values from the normal distribution
fig, ax = plt.subplots()
ax.hist(x)
```

```python
#Suplot option 1
fig, ((ax0, ax1), (ax2, ax3))  = plt.subplots(nrows = 2,
                               ncols = 2,
                               figsize = (10,5))
#Plot to each different plot
ax0.plot(x, x/2);
ax0.set(title="Line plot")
ax1.scatter(np.random.random(10), np.random.random(10));
ax1.set(title = "Scatter Plot")
ax2.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax2.set(title = "Bar Graph")
ax3.hist(np.random.randn(1000))
ax3.set(title = "Histogram")
```

```python
#Suplot option 2
fig, ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10,5))
#Plot to each different index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

### pyplot  vs MPL O.O method
* when plotting something quick, its okay to use the pyplot method
* for advanced plots, use Matplotlib Object oriented method. (below)

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
#pyplot method
over_50.plot(kind= "scatter",
            x="age",
            y="chol",
            c="target")
```

```python
# OO method + pyplot 
fig, ax= plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax= ax)
# ax.set_xlim([45,100])
```

```python
#OO method
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"],y = over_50["chol"],
                    c=over_50['target'])
#customize
ax.set(title= "Heart Disease Cholesterol Levels",
      xlabel= "Age",
      ylabel= "Cholesterol")
ax.legend(*scatter.legend_elements(), title= "Target")
#horizontal line at the mean
ax.axhline(over_50["chol"].mean(),
          linestyle=":")
```

```python
#subplot of chol age and thalach
fig, (ax0,ax1) = plt.subplots(nrows = 2,
                             ncols= 1,
                             figsize=(10,10),
                             sharex= True)
#add data
scatter0 = ax0.scatter(x= over_50['age'], y=over_50["chol"], c=over_50["target"])
scatter1 = ax1.scatter(x= over_50["age"], y= over_50["thalach"], c= over_50['target'])

#customize
ax0.set(title="Cholesterol levels compared to age",
#        xlabel= "Age",
       ylabel= "Cholesterol")
ax0.legend(*scatter0.legend_elements(), title= "Target")
ax1.set(title="Cholesterol levels compared to heart rate",
       xlabel= "Age",
       ylabel= "Max. Heart rate")
ax1.legend(*scatter1.legend_elements(), title= "Target")
#horixontal line
ax0.axhline(over_50["chol"].mean(), linestyle = ":")
ax1.axhline(over_50["thalach"].mean(), linestyle= "--")

#add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize= 16, fontweight= 'bold')
```

```python
plt.style.use('seaborn-whitegrid')
#OO method
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"],y = over_50["chol"],
                    c=over_50['target'],
                    cmap= "summer")
#customize
ax.set(title= "Heart Disease Cholesterol Levels",
      xlabel= "Age",
      ylabel= "Cholesterol")
ax.legend(*scatter.legend_elements(), title= "Target")
#horizontal line at the mean
ax.axhline(over_50["chol"].mean(),
          linestyle=":")
```

```python
#Customizing y and x axes limitations
#subplot of chol age and thalach
fig, (ax0,ax1) = plt.subplots(nrows = 2,
                             ncols= 1,
                             figsize=(10,10),
                             sharex= True)
#add data
scatter0 = ax0.scatter(x= over_50['age'], y=over_50["chol"], c=over_50["target"], cmap= "winter")
scatter1 = ax1.scatter(x= over_50["age"], y= over_50["thalach"], c= over_50['target'], cmap="winter")

#customize
ax0.set(title="Cholesterol levels compared to age",
#        xlabel= "Age",
       ylabel= "Cholesterol")
ax0.legend(*scatter0.legend_elements(), title= "Target")
ax0.set_xlim([50,80])
ax1.set(title="Cholesterol levels compared to heart rate",
       xlabel= "Age",
       ylabel= "Max. Heart rate")
ax1.set_ylim([60,200])
ax1.legend(*scatter1.legend_elements(), title= "Target")
#horixontal line
ax0.axhline(over_50["chol"].mean(), linestyle = ":")
ax1.axhline(over_50["thalach"].mean(), linestyle= "--")

#add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize= 16, fontweight= 'bold')
```
